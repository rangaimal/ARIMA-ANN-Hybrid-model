```python
train_size = int(df.shape[0]*0.8)
train, test = df.ASPI[0:train_size], df.ASPI[train_size:]
print(train.shape)
```

```python
data = train
```

```python
predict = []
for t in test:
    model = ARIMA(data, order=(4,1,0))

    model_fit = model.fit()
    y = model_fit.forecast()
    # up to this this is auto regression model. y is combination of 3 arrays. we get forecast values
    print(y[0][0])
    # store y values in predict model
    predict.append(y[0][0])
    data = np.append(data, t)
    data = pd.Series(data)
   
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, median_absolute_error, mean_squared_log_error
```

```python
MAE = mean_absolute_error(test, predict)
MAE
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
MAPE = mean_absolute_percentage_error(test, predict)
```

```python
print(MAPE)
```

```python
plt.plot(test)
plt.plot(predict, color='red')
plt.show()
```

```python
from math import sqrt
rmse = sqrt(mean_squared_error(test, predict))
rmse
```
