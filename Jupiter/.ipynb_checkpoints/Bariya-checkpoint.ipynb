```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
import tensorflow as tf
from keras.models import Sequential, load_model
```

```python
from numpy.random import seed
seed(1)
tf.random.set_seed(1)
seed = 1
np.random.seed(seed)
```

```python
dataset = pd.read_csv(r'C:\Users\Acer\Downloads\Finalized_V2.csv')
dataset.head()
```

```python
X = dataset.iloc[:, 1:17].values
y = dataset.iloc[:, 17].values
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
classifier = Sequential()
classifier.add(Dense(output_dim = 9, init = 'uniform', activation = 'relu', input_dim = 16))
classifier.add(Dense(output_dim = 9, init = 'uniform', activation = 'relu'))
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
classifier.save('stock_prediction4.h5')
```

```python
classifier = load_model('stock_prediction4.h5')
```

```python
y_pred = classifier.predict(X_test)
y_pred=(y_pred>0.5)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
print ("The prediction accuracy is: " , metrics.accuracy_score(y_test, y_pred))
print("The mean squared error rate is: ",metrics.mean_squared_error(y_test,y_pred))
print("The absolute squared error rate is: ",metrics.mean_absolute_error(y_test,y_pred))
print("The precision score is: ",metrics.precision_score(y_test,y_pred))
print("The recall score is: ",metrics.recall_score(y_test,y_pred))
print("The f1 score is: ",metrics.f1_score(y_test,y_pred))
print("The confusion matrix is: ",metrics.confusion_matrix(y_test,y_pred))
```

```python
kf = KFold(n_splits=7)

def get_score(model, X_train, X_test, y_train, y_test):
        model.fit(X_train, y_train)
        return model.evaluate(X_test, y_test)

scores_SVM=[]

for train_index,test_index in kf.split(X,y):
    print("Train Index: ", train_index, "\n")
    print("Test Index: ", test_index)

    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    scores_SVM.append(get_score(classifier, X_train, X_test, y_train, y_test))
    print(scores_SVM)
```
